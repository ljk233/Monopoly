# Testing Monopoly v2

## Testing the `Account` class.

```python
from Monopoly import Account, Player
```

```python
# declare and initialise Account object
a1: Account = Account()
```

```python
a1.describe()
```

```python
assert a1.get_balance() == 500
```

```python
a1.set_balance(300)
assert a1.get_balance() == 300
```

```python
a1.credit(100)
assert a1.get_balance() == 400
```

```python
a1.debit(400)
assert a1.get_balance() == 0
```

```python
assert a1.debit(400) == False
```

## Testing `Player`

```python
p1: Player = Player()
p2: Player = Player()
```

```python
p1.assign_to("Player 1", "Dog")
p2.assign_to("Player 1", "Duck")
```

```python
p1.describe()
```

```python
p2.describe()
```

```python
p1.account_credit(100)
assert p1.get_account_balance() == 600
```

```python
p1.account_debit(100)
assert p1.get_account_balance() == 500
```

```python
assert p1.account_transfer(p2, 500) == True
```
