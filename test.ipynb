Testing the `Player` class.

```python
from Monopoly.Player import Player
```

Declare and initialise two players.

```python
p1: Player = Player()
p2: Player = Player()
```

Test the **getters** and **setters.**

```python
# name
p1.set_name("Player 1")
p1.get_name()
```

```python
p1.set_piece("Dog")
p1.get_piece()
```

```python
p1.set_account_balance(500)
p1.get_account_balance()
```

Test the class method `ROLL_DICE()`

```python
p1.ROLL_DICE()
```

Test `describe()` method.

```python
p1.describe()
```

Test `assign_to()` method.

```python
p2.assign_to("Player 2", "Car", 500)
p2.describe()
```

Finally, test `is_same_player_as()`.

```python
p1.is_same_player_as(p2)  # expected, false
```
